```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ============================
# Step 1: Dataset
# ============================
# Student data: Hours studied vs Test grades
students = ["Ryan", "Kristen", "James", "Runa", "Kate"]
hours_studied = np.array([3, 5, 7, 9, 11]).reshape(-1, 1)  # X
# When you use .reshape(-1, 1) on a NumPy array, it changes the shape of the array so it fits what many libraries (like scikit-learn) expect.
grades = np.array([75, 80, 85, 90, 95])                   # Y

# ============================
# Step 2: Linear Regression
# ============================
model = LinearRegression()
model.fit(hours_studied, grades)

# Get slope (m) and intercept (b)
m = model.coef_[0]
b = model.intercept_
print(f"Regression equation: y = {m:.2f}x + {b:.2f}")

# ============================
# Step 3: Predictions
# ============================
# Predict grades based on the regression line
predicted_grades = model.predict(hours_studied)

# ============================
# Step 4: Plotting
# ============================
# Generate x-values for smooth line
x_fit = np.linspace(0, 12, 100).reshape(-1, 1)
y_fit = model.predict(x_fit)

# Plot original data points
plt.scatter(hours_studied, grades, color="blue", label="Student Data")
for i, student in enumerate(students):
    plt.text(hours_studied[i], grades[i]+0.5, student)

# Plot regression line
plt.plot(x_fit, y_fit, color="red", label=f"Regression Line: y={m:.2f}x+{b:.2f}")

# Chart formatting
plt.title("Regression Analysis: Hours Studied vs Test Grades")
plt.xlabel("Hours Studied")
plt.ylabel("Test Grade")
plt.legend()
plt.grid(True)
plt.show()

# ============================
# Step 5: Print Predictions
# ============================
for i in range(len(students)):
    print(f"Student {students[i]}: studied {hours_studied[i][0]} hours, actual grade {grades[i]}, predicted grade {predicted_grades[i]:.2f}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# AND dataset
# Input (X) and Output (y)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [0], [0], [1]])  # AND logic

# Create a simple perceptron model
model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))  # 1 neuron, sigmoid activation

# Compile the model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=100, verbose=1)

# Evaluate the model
print("Evaluation:")
print(model.predict(X).round())  # Round to get binary output
```
