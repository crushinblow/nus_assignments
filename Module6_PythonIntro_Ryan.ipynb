Data structures in python
- list -> changeable (shopping list )
- tuple -> unchangeable (emergency numbers)
- dictionary -> phonebook (name: Number)
- set -> collection of uniue items

```python
fruits = ['apple', 'mango','banana']
print(fruits)
print(len(fruits))
fruits.append('orange')
print(fruits)
print(len(fruits))
fruits.pop()
print(fruits)
fruits.remove('apple')
print(fruits)
fruits.count('apple')

for i in fruits:
  print(i)

for i in range(0,len(fruits)):
  print(fruits[i])
```
