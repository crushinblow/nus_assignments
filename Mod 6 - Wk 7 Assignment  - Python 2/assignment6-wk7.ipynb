```python
import numpy as np

"""
Assignment 7
Qn 1:
Part 1a: Create a 2D NumPy array of shape (3, 3) with float64 elements
- Create a nested list structure representing a 3x3 matrix (2D - rows and columns)
- Explicitly specify dtype=np.float64 for 64-bit floating point

"""
arr = np.array([[1.5, 2.3, 3.1],
                [4.2, 5.0, 6.7],
                [7.8, 8.4, 9.1]], dtype=np.float64)


"""
Part 1b: Element-wise addition with scalar 10
- NumPy automatically broadcasts and adds it to every element of the array
to match the array shape
- The operation is performed element-wise (so each element gets +10)
"""
result_add = arr + 10

"""
Part 1c: Element-wise multiplication with another 3x3 array
- Create a second array with compatible dimensions (3x3)
- The multiplication is performed element-by-element
- Position [0,0] in arr multiplies with position [0,0] in arr2, etc.
"""
arr2 = np.array([[1.7, 0.8, 1.5],
                 [2.3, 2.2, 1.7],
                 [2.6, 42.0, 6.2]])
result_mul = arr * arr2

"""
Part 1d: Print array attributes for both result arrays
- dtype: will show the data type of array elements (should be float64 for both)
- shape: will show the dimensions (rows, columns) as a tuple
- size: will show total number of elements (rows × columns)
"""
print("Addition result array properties:")
print("dtype:", result_add.dtype)    # Data type of elements
print("shape:", result_add.shape)    # Array dimensions (3, 3)
print("size:", result_add.size)      # Total elements (9)

print("\nMultiplication result array properties:")
print("dtype:", result_mul.dtype)    # Should match input types
print("shape:", result_mul.shape)    # Should remain (3, 3)
print("size:", result_mul.size)      # Should remain 9

# ===================================
# 2. Data Cleaning & Missing Values
# ===================================
import pandas as pd

"""
Part 2a: Create DataFrame with missing values
- Create a dictionary where keys become column names
- np.nan represents missing/undefined values (Not a Number)
- The DataFrame will automatically handle these as missing values
"""
data = {'A': [1, 2, np.nan, 4],       # Column A with one missing value
        'B': [5, np.nan, np.nan, 8],   # Column B with two missing
        'C': [9, 10, 11, np.nan]}      # Column C with one missing
df = pd.DataFrame(data) #Pandas function to create a tabular data structure (like a spreadsheet or SQL table).

"""
Part 2b: Identify missing values
- isna() and isnull() are aliases (work exactly the same)
- Returns a boolean DataFrame where True indicates missing values
- Useful for counting or locating missing data
"""
missing_values = df.isna() #df.isna() is a Pandas DataFrame/Series method that checks for missing values (NaN, None, or NA).
print("\nMissing values detection:")
print(missing_values)

"""
Part 2c: Drop rows with any missing data
- dropna() by default removes rows with any missing values
- Important: This creates a new DataFrame, doesn't modify original
- Alternative: Use inplace=True to modify original
"""
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with missing values:")
print(df_dropped)

"""
Part 2d: Fill missing values
- fillna() provides multiple strategies for handling missing data
- Simple replacement: fill with a constant value (0 in first example)
- Statistical replacement: fill with column mean (preserves some distribution)
- Other options: median, mode, forward fill, etc.
"""
print("\nDataFrame with missing values filled with 0:")
df_filled_zero = df.fillna(0)
print(df_filled_zero)

print("\nDataFrame with missing values filled with column means:")
df_filled_mean = df.fillna(df.mean())
print(df_filled_mean)

# ======================
# 3. Matplotlib Visualizations
# ======================
import matplotlib.pyplot as plt #import the pyplot module from Matplotlib

"""
Part 3a: Simple line plot with generated data
- We create time values from 0 to 10 in 0.5 increments
- Compute sine values for demonstration (creates wave pattern)
- Basic plot() connects points with lines
"""
time = np.arange(0, 10, 0.5)  # Generate time values from 0-10 in 0.5 steps with Numpy method np
values = np.sin(time)          # Compute sine of each time value

plt.figure(figsize=(8, 4))     # Create new figure with specified size
plt.plot(time, values)         # Basic line plot

"""
Part 3b: Add labels and title
- xlabel: Describes x-axis (typically independent variable)
- ylabel: Describes y-axis (typically dependent variable)
- title: Overall description of the plot
"""
plt.xlabel('Time (seconds)')    # Label for x-axis
plt.ylabel('Value')             # Label for y-axis
plt.title('Value Change Over Time')  # Plot title
plt.grid(True)                  # Add grid lines for better readability
plt.show()                      # Display the plot

"""
Part 3c: Customize line plot appearance
- color: Changes line color (can use names or hex codes)
- marker: Adds symbols at data points ('o' for circles)
- linestyle: Changes line pattern ('--' for dashed)
- linewidth: Controls line thickness (in points)
"""
plt.figure(figsize=(8, 4))
plt.plot(time, values,
         color='red',          # Set line color to red
         marker='o',           # Add circular markers at data points
         linestyle='--',       # Use dashed line style
         linewidth=2)          # Make line thicker
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Customized Value Change Over Time')
plt.grid(True)
plt.show()

"""
Part 3d: Scatter plot of age vs income
- Generate random but realistic age and income data
- scatter() shows individual points without connecting lines
- Useful for visualizing relationships/correlations
"""
np.random.seed(42)  # For reproducible random numbers
age = np.random.randint(20, 60, size=50)         # Random ages 20-60
income = np.random.randint(20000, 100000, size=50)  # Random incomes

plt.figure(figsize=(8, 5))
plt.scatter(age, income,
           color='blue',       # Point color
           alpha=0.6)          # Slightly transparent points
plt.xlabel('Age')              # x-axis label
plt.ylabel('Income ($)')       # y-axis label with units
plt.title('Age vs Income Relationship')  # Descriptive title
plt.grid(True)
plt.show()
```
